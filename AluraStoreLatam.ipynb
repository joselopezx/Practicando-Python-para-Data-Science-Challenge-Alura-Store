### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda, tienda2, tienda3, tienda4]

tienda.head()
```

```python
ingresoTotal1= int(tienda['Precio'].sum())
ingresoTotal1
```

```python
ingresoTotal3 = int(tienda3['Precio'].sum())
ingresoTotal3
```

```python
ingresoTotal4 = int(tienda4['Precio'].sum())
ingresoTotal4
```

```python
productosCategoria = tienda['Categoría del Producto'].value_counts()
productosCategoria
```

```python
productosCategoria.plot(
    kind='bar',
    figsize=(8, 4),
    color='skyblue',
    alpha=0.6  # transparencia
)

plt.title('Distribución de productos por categoría - Tienda 1')
plt.xlabel('Categoría del producto')
plt.ylabel('Cantidad de productos')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

#1. Análisis de facturación



```python
ingresosTotales = []
for tienda in tiendas:
  ingresosTotales.append(int(tienda['Precio'].sum()))
print(ingresosTotales)
```

# 2. Ventas por categoría

```python
productosCategoriaTiendas = []
for tienda in tiendas:
  productosCategoriaTiendas.append(tienda['Categoría del Producto'].value_counts())
productosCategoriaTiendas
```

# 3. Calificación promedio de la tienda


```python
promedios = []
for tienda in tiendas:
  promedios.append(float(tienda['Calificación'].mean()))
promedios
```

# 4. Productos más y menos vendidos

```python
for i, tiendaProductos in enumerate(productosCategoriaTiendas, start=1):
  print(f"Tienda {i} : {tiendaProductos.idxmax()} con {tiendaProductos.max()}")
```

```python
for i, tiendaProductos in enumerate(productosCategoriaTiendas, start=1):
    categoria_mas_popular = tiendaProductos.idxmax()
    cantidad = tiendaProductos.max()
    print(f"Tienda {i}: {categoria_mas_popular} ({cantidad} productos)")
```

```python
for i in productosCategoriaTiendas:
  i.plot(kind='pie', figsize=(10,4.5), autopct='%1.1f%%')
  plt.show()
```

```python
productosCategoriaTiendas[1].plot(kind='bar', figsize=(6,3), color='purple', xlabel='Categoria', ylabel='Cantidad')
```

# 5. Envío promedio por tienda

```python
envioPromedios = []
for tienda in tiendas:
  envioPromedios.append(float(tienda['Costo de envío'].mean()))
print(envioPromedios)
envioPromedios = pd.DataFrame(envioPromedios)
envioPromedios.plot(kind='area', figsize=(6,3), color='purple', xlabel='tienda', ylabel='promedio', alpha=0.6)
```
